```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
def load_data():
    with open('X_train.npy', 'rb') as f:
        X_train = np.load(f)
    with open('Y_train.npy', 'rb') as f:
        Y_train = np.load(f)
    with open('X_test.npy', 'rb') as f:
        X_test = np.load(f)
    with open('Y_test.npy', 'rb') as f:
        Y_test = np.load(f)
    f.close()
    return X_train,X_test,Y_train,Y_test
```

```python
def tanh(z):
    t = (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))
    return t
def tanh_dash(z):
    t = 1 - np.square(tanh(z))
    return t
def ReLU(x):
    return x * (x > 0)
def dReLU(x):
    return 1. * (x > 0)
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
```

```python
def initalise():
    W1 = np.random.randn(100,8*8*3)*0.01
    b1 = np.zeros((100,1))
    W2 = np.random.randn(1,100)*0.01
    b2 = np.zeros((1,1))
    return W1,b1,W2,b2
```

```python
def NN_train(W1,b1,W2,b2,m,X_train,Y_train,g1,g1_dash,g2,learning_rate):
    #forward propagation
    Z1 = np.dot(W1,X_train) + b1
    A1 = g1(Z1)
    Z2 = np.dot(W2,A1) + b2
    A2 = g2(Z2)
    #backword propagation
    dZ2 = A2 - Y_train
    dW2 = (1/m)*np.dot(dZ2,A1.T)
    db2 = (1/m)*np.sum(dZ2,axis=1,keepdims=True)
    dZ1 = np.dot(W2.T,dZ2)*g1_dash(Z1)
    dW1 = (1/m)*np.dot(dZ1,X_train.T)
    db1 = (1/m)*np.sum(dZ1,axis=1,keepdims=True)
    #update
    W2 = W2 - learning_rate*dW2
    b2 = b2 - learning_rate*db2
    W1 = W1 - learning_rate*dW1
    b1 = b1 - learning_rate*db1
    return W1,b1,W2,b2
```

```python
def NN_accuracy(W1,b1,W2,b2,X_test,Y_test,g1,g2):
    #preditction
    Z1 = np.dot(W1,X_test) + b1
    A1 = g1(Z1)
    Z2 = np.dot(W2,A1) + b2
    A2 = g2(Z2)
    #accuracy
    count = 0
    m_test = A2.size
    for i in range (m_test):
        if A2[0][i]<=0.5:
            out = 0
        else:
            out = 1
        if out == Y_test[0][i]:
            count = count + 1
    acc = (count/m_test)*100
    return acc
```

```python
X_train,X_test,Y_train,Y_test = load_data()
W1,b1,W2,b2 = initalise()
learning_rate = 0.1
epoch = 1000
m = 8000
for i in range (epoch):
    W1,b1,W2,b2 = NN_train(W1,b1,W2,b2,m,X_train,Y_train,ReLU,dReLU,sigmoid,learning_rate)
    acc = NN_accuracy(W1,b1,W2,b2,X_test,Y_test,ReLU,sigmoid)
    print(str(acc)+"%")
```
